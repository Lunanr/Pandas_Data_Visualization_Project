```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('iris.csv',index_col='Id')
```

```python
data.info()
```

```python
data.head()
```

## 1. Draw a line plot for all the numerical columns

```python
data.plot()
#line plot is a default plot in this function
plt.show()
```

## 2. Change size of a figure & font: Figure size: The width of 15 and height of 10, Font size : 10

```python
data.plot(figsize=(15,10),fontsize=15)
plt.show()
```

## 3. Make separate subplots for each column

```python
data.plot(subplots=True,figsize=(15,10),fontsize=15)
#Subplots for make a all of graph from the dataset
plt.show()
```

## 4. Share x-axis values with other graphs

```python
data.plot(subplots=True,figsize=(15,10),fontsize=15,sharex=False)
#sharex for share x-axis values
plt.show()
```

## 5. Share Y-axis values with other graphs

```python
data.plot(subplots=True,figsize=(15,10),fontsize=15,sharex=False,sharey=True)
plt.show()
```

## 6. Draw the line plot only for petal width column

```python
data.columns
```

```python
data['PetalWidthCm'].plot()
plt.show()
```

```python
data[['PetalLengthCm','PetalWidthCm']].plot()
plt.show()
```

## 7. Add the title and labels to X-axis & Y-axis

```python
data['PetalWidthCm'].plot(title="Iris Petal Width",
                         xlabel="Indexes",ylabel="Petal Width in Cm")
plt.show()
```

## 8. Draw the line plot for petal length & petal width columns (Also Change Line Style)

```python
data['PetalLengthCm'].plot(style='--')
plt.show()
```

```python
data[['PetalLengthCm','PetalWidthCm']].plot(style=['--','-.'])
plt.show()
```

## 9. Draw the line plot for petal length & petal width column (Change limit of X and Y axis)

```python
data[['PetalLengthCm','PetalWidthCm']].plot(style=['--','-.'],
                                           xlim=(10,140),
                                           ylim=(1,10))
plt.show()
```

## 10. xticks and yticks

```python
x=[x for x in range(0,140,10)]
y=[y for y in range(0,10,3)]
```

```python
## xticks,yticks : to customize x ticks paramater
data[['PetalLengthCm','PetalWidthCm']].plot(style=['--','-.'],
                                           xticks=x,
                                           yticks=y)
plt.show()
```

## 11. Area plot

```python
data.plot(kind='area',alpha=0.5)
#alpha for make transpracy in area plot
plt.show()
```
